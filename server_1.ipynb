```python
import socket
import threading


# Sunucu adresi ve port numarası
HOST = '127.0.0.1'  # localhost
PORT = 7546

# Bağlı istemcilerin listesi ve sıcaklık, hız, voltage, ve temp verilerini tutacak liste
clients = []
temperature_val = []
speed_val = []
voltage_val = []
custom_val = []
#burda ekleme yaptım
custom_val2= []
custom_val3= []

start = ["start"]
stop = ["stop"]

def write_temperature(message):
    new_temperature = int(message.split("*")[2])
    temperature_val.clear()  # Tüm önceki değerleri sil
    temperature_val.append(new_temperature)
    print("Sıcaklık verisi güncellendi:", new_temperature)

def read_temperature(client_socket):
    if temperature_val:
        client_socket.send(str(temperature_val[-1]).encode())
    else:
        client_socket.send("Sıcaklık verisi bulunamadı.".encode())

def write_speed(message):
    try:
        new_speed = int(message.split("*")[2])
        speed_val.clear()  # Tüm önceki değerleri sil
        speed_val.append(new_speed)
        print("Hız verisi güncellendi:", new_speed)
    except ValueError:
        print("Hatalı hız değeri:", message.split("*")[2])
        print("Hatalı hız değeri:", message)

def read_speed(client_socket):
    if speed_val:
        client_socket.send(str(speed_val[-1]).encode())
    else:
        client_socket.send("Hız verisi bulunamadı.".encode())

def write_voltage(message):
    new_voltage = int(message.split("*")[2])
    voltage_val.clear()  # Tüm önceki değerleri sil
    voltage_val.append(new_voltage)
    print("Voltage verisi güncellendi:", new_voltage)

def read_voltage(client_socket):
    if voltage_val:
        client_socket.send(str(voltage_val[-1]).encode())
    else:
        client_socket.send("Voltage verisi bulunamadı.".encode())

def write_custom(message):
    new_value = int(message.split("*")[2])
    custom_val.clear()  # Tüm önceki değerleri sil
    custom_val.append(new_value)
    print("Özel değer güncellendi:", new_value)

def read_custom(client_socket):
    if custom_val:
        client_socket.send(str(custom_val[-1]).encode())
    else:
        client_socket.send("Özel değer bulunamadı.".encode())
#burda ekleme yaptım
def write_custom2(message):
    new_value2 = int(message.split("*")[2])
    custom_val2.clear()  # Tüm önceki değerleri sil
    custom_val2.append(new_value2)
    print("Özel değer güncellendi:", new_value2)

def read_custom2(client_socket):
    if custom_val2:
        client_socket.send(str(custom_val2[-1]).encode())
    else:
        client_socket.send("Özel değer bulunamadı.".encode())

def write_custom3(message):
    new_value3 = int(message.split("*")[2])
    custom_val3.clear()  # Tüm önceki değerleri sil
    custom_val3.append(new_value3)
    print("Özel değer güncellendi:", new_value3)

def read_custom3(client_socket):
    if custom_val3:
        client_socket.send(str(custom_val3[-1]).encode())
    else:
        client_socket.send("Özel değer bulunamadı.".encode())


def handle_client(client_socket, self=None):
    while True:
        try:
            # İstemciden veri al
            data = client_socket.recv(1024)
            if not data:
                # Veri alınmadıysa bağlantıyı kapat
                print("İstemci bağlantısı kapatıldı.")
                clients.remove(client_socket)
                client_socket.close()
                break
            else:
                # Gelen veriyi işle
                message = data.decode()
                if message.startswith("write*1"):
                    if start:
                        # start listesindeki ilk elemanı al
                        start_command = start[0]
                        # İlk bağlı istemciye gönder
                        if clients:
                            first_client_socket = clients[0]
                            first_client_socket.send(start_command.encode())
                            print("write*1 komutu gönderildi:", start_command)
                        else:
                            print("Bağlı istemci yok, komut gönderilemedi.")
                    else:
                        print("Başlatma komutu yok, işlem yapılamadı.")
                elif message.startswith("write*2"):
                    if stop:
                        # stop listesindeki ilk elemanı al
                        stop_command = stop[0]
                        # İlk bağlı istemciye gönder
                        if clients:
                            first_client_socket = clients[0]
                            first_client_socket.send(stop_command.encode())
                            print("write*2 komutu gönderildi:", stop_command)
                        else:
                            print("Bağlı istemci yok, komut gönderilemedi.")
                    else:
                        print("Durdurma komutu yok, işlem yapılamadı.")
                elif message.startswith("write*4*"):
                    write_temperature(message)
                elif message.startswith("read*4"):
                    read_temperature(client_socket)
                elif message.startswith("write*5*"):
                    write_speed(message)
                elif message.startswith("read*5"):
                    read_speed(client_socket)
                elif message.startswith("write*6*"):
                    write_voltage(message)
                elif message.startswith("read*6"):
                    read_voltage(client_socket)
                elif message.startswith("write*7"):
                    write_custom(message)
                elif message.startswith("read*7"):
                    read_custom(client_socket)
                    #burda ekleme yaptım
                elif message.startswith("write*8"):
                    write_custom2(message)
                elif message.startswith("read*8"):
                    read_custom2(client_socket)
                elif message.startswith("write*9"):
                    write_custom3(message)
                elif message.startswith("read*9"):
                    read_custom3(client_socket)
                else:
                    print("Bilinmeyen komut:", message)
        except socket.error as e:
            # Hata durumunda istemciyi listeden çıkar ve bağlantıyı kapat
            print(e)
            clients.remove(client_socket)
            client_socket.close()
            break

def start_server():
    # IPv4 ve TCP soketi oluştur
    server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server.bind((HOST, PORT))
    server.listen()

    print("Sunucu dinleniyor...")

    while True:
        # İstemci bağlantıları kabul et
        client_socket, client_address = server.accept()
        print(f"{client_address} adresinden istemci bağlandı.")

        # Bağlı istemciler listesine ekle
        clients.append(client_socket)

        # İstemci işlemleri için yeni bir thread başlat
        client_handler = threading.Thread(target=handle_client, args=(client_socket,))
        client_handler.start()

if __name__ == "__main__":
    start_server()
```
