```python
import socket
import threading
import random
import time

class Client:
    def __init__(self, host, port):
        self.host = host
        self.port = port
        self.client_socket = None
        self.send_data = False
        self.lock = threading.Lock()

    def connect_to_server(self):
        self.client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        try:
            self.client_socket.connect((self.host, self.port))
            print("Sunucuya bağlandı.")
            self.start_sending_data_thread = threading.Thread(target=self.start_sending_data)
            self.start_sending_data_thread.start()
        except Exception as e:
            print("Sunucuya bağlanırken bir hata oluştu:", e)

    def disconnect_from_server(self):
        if self.client_socket:
            self.client_socket.close()
            print("Sunucudan bağlantı kesildi.")
            self.stop_sending_data()
        else:
            print("Hata: Bağlantı henüz kurulmamış.")

    def send_command(self, command):
        if self.client_socket:
            try:
                self.client_socket.send(command.encode())
            except Exception as e:
                print("Sunucuya komut gönderirken bir hata oluştu:", e)

    def start_sending_data(self):
        self.send_data = False  # Başlangıçta, sunucudan başlatma komutu gelene kadar False olarak ayarlanır
        while not self.send_data:
            try:
                data = self.client_socket.recv(1024).decode()  # Sunucudan veri al
                if data == "start":
                    print("Sunucu, veri göndermeyi başlatmamızı istedi.")
                    self.send_data = True  # Veri göndermeyi başlatmak için send_data True olarak ayarlanır
                elif data == "stop":
                    print("Sunucu, veri göndermeyi durdurmayı istedi.")
                    self.send_data = False  # Veri göndermeyi durdurmak için send_data False olarak ayarlanır
                    self.stop_sending_data()  # Veri göndermeyi durdurmak için stop_sending_data metodunu çağır
                    return  # Fonksiyondan çıkmak için return kullanılır
            except Exception as e:
                print("Sunucudan veri alınırken bir hata oluştu:", e)
                self.stop_sending_data()

        # Veri göndermeye başla
        while self.send_data:
            try:
                temperature_value = random.randint(0, 100)
                speed_value = random.randint(90, 180)
                voltage_value = random.randint(120, 360)
                custom_value = random.randint(0, 105)  # Yeni eklenen özel değer
                #burda ekleme yaptım
                custom_value2= random.randint(0,100)
                custom_value3= random.randint(0,100)

                # Rasgele veriyi sunucuya gönder
                self.send_command(f"write*4*{temperature_value}")
                print(f"Sunucuya gönderilen komut: write*4*{temperature_value}")
                time.sleep(0.0001)  # Kısa bir süre bekle

                self.send_command(f"write*5*{speed_value}")
                print(f"Sunucuya gönderilen komut: write*5*{speed_value}")
                time.sleep(0.0001)  # Kısa bir süre bekle

                self.send_command(f"write*7*{custom_value}")  # Yeni eklenen özel değer
                print(f"Sunucuya gönderilen komut: write*7*{custom_value}")  # Yeni eklenen özel değer
                time.sleep(0.0001)  # Kısa bir süre bekle
                #burda ekleme yaptım
                self.send_command(f"write*8*{custom_value2}")  # Yeni eklenen özel değer
                print(f"Sunucuya gönderilen komut: write*8*{custom_value2}")  # Yeni eklenen özel değer
                time.sleep(0.0001)  # Kısa bir süre bekle
                self.send_command(f"write*9*{custom_value3}")  # Yeni eklenen özel değer
                print(f"Sunucuya gönderilen komut: write*9*{custom_value3}")  # Yeni eklenen özel değer
                time.sleep(0.0001)  # Kısa bir süre bekle

                self.send_command(f"write*6*{voltage_value}")
                print(f"Sunucuya gönderilen komut: write*6*{voltage_value}")
                time.sleep(4)  # Veri gönderme arasında 4 saniye bekle
            except Exception as e:
                print("Veri gönderirken bir hata oluştu:", e)
                self.stop_sending_data()

    def stop_sending_data(self):
        self.send_data = False

if __name__ == "__main__":
    client = Client('127.0.0.1', 7546)
    client.connect_to_server()
```
